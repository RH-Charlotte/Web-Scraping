# Web Scraping - Amazon Product Reviews 

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page = requests.get('https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews')
soup = bs(page.text)
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0)
cust_name
```

```python
cust_name.pop(0)
cust_name.pop(4)
cust_name.pop(7)
cust_name.pop(8)
cust_name
```

```python
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
len(rate)
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
rate
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
cust_name
review_title
rate
review_content
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
df['Customer Name']=cust_name
```

```python
df
```

```python
df['Review title']=review_title
df['Ratings']=rate
df['Reviews']=review_content
```

```python
df
```

```python
df['#Rating'] = df['Ratings'].str.slice(0, 3)
df
```

```python
df['#Rating'] = pd.to_numeric(df['#Rating'])
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['#Rating'])
plt.ylabel('Ratings')
plt.yticks(range(1,6))
plt.xticks(range(1,6))
plt.title('Rating Distribution')
plt.grid(True)
```
